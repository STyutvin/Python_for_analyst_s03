```python
# Импортируем pandas

import pandas as pd
```

```python
# Считываем данные и выводим первые пять строк

houses = pd.read_csv('./kc_house_data.csv')
houses.head()
```

1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.
Если реновации дома не было, то в новом признаке поставьте 0.

*Возможно есть более изящное решение, но, чтобы не запутаться, сделаю так, как мне представляется понятным.*

```python
# Создадим пустой массив diff=[], и в него будем записывать результаты для нового признака delta_renovated со следующим условием

diff=[]
for i, row in houses.iterrows():
  if row['yr_renovated']-row['yr_built']>0:
    diff.append(row['yr_renovated']-row['yr_built'])
  else:
    diff.append(0)

# Теперь добавим полученный массив в DataFrame и выведем результат для проверки

houses['delta_renovated']=diff
houses
```

1.3 Создайте признаки года продажи, месяца продажи.

*Поскольку данные о продажах неизвестны, то столбцы с годом и месяцем продажи заполним нулями.*

```python
# Для года продажи возьмем обозначение yr_sold, а для месяца month_sold

houses[['yr_sold','month_sold']]=0
houses
```

1.4 Удалите признаки date, zipcode, lat, long.

```python
houses.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
houses.head()
```
